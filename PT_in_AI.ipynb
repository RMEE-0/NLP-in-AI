```
!pip install nltk spacy textblob gradio
!python -m textblob.download_corpora
!python -m spacy download en_core_web_sm

import nltk
nltk.download("punkt")

import spacy
from textblob import TextBlob
import gradio as gr

nlp = spacy.load("en_core_web_sm")

# -------------------------------
# NLP FUNCTION (must be BEFORE Gradio)
# -------------------------------
def nlp_analyze(text):
    doc = nlp(text)

    tokens = [token.text for token in doc]
    pos_tags = [(token.text, token.pos_) for token in doc]
    named_entities = [(ent.text, ent.label_) for ent in doc.ents]

    blob = TextBlob(text)
    sentiment = blob.sentiment

    output = f"""
### NLP Analysis Results

---

#### Tokens
{tokens}

---

#### Part-of-Speech (POS) Tags
{pos_tags}

---

#### Named Entities (NER)
{named_entities}

---

#### Sentiment
- Polarity: {sentiment.polarity}
- Subjectivity: {sentiment.subjectivity}

---
"""
    return output

# -------------------------------
# GRADIO APP
# -------------------------------
interface = gr.Interface(
    fn=nlp_analyze,
    inputs=gr.Textbox(lines=4, placeholder="Enter text here..."),
    outputs="markdown",
    title="NLP Analyzer App",
    description="Tokenization, POS Tagging, Named Entity Recognition, and Sentiment Analysis."
)

interface.launch()
```
